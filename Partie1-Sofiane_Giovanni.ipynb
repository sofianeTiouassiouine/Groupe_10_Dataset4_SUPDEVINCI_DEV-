```python
def describe1(data): #cette fonction permet de décrire toutes les variables quantitatives
    print(data.describe())
    print("---------------")
describe1(tab)

def describe2(data, colonne):#cette fonction permet de décrire une variable quantitatives
    print(data[colonne].describe())
describe2(tab, 'Education')
```

```python
tab.isnull().sum(axis = 0)#compte les occurences NaN pour colonne
```

```python
def countNancolonne(data): #cette fonction permet de compter les occurences NaN pour chaque colonne
    print(data.isnull().sum(axis = 0))
    print("---------------")
countNancolonne(tab)

def countNanligne(data):#cette fonction permet de compter les occurences NaN pour chaque colonne
    print(data.isnull().sum(axis = 1))
countNanligne(tab)
```

```python
def affichelignehaut(data, nombre): #cette fonction d'afficher avec personnalisation le nombre de ligne du haut du tableau
    print(data.head(nombre))
    print("---------------")
affichelignehaut(tab, 10)

def affichelignebas(data, nombre):#cette fonction d'afficher avec personnalisation le nombre de ligne du bas du tableau
    print(data.tail(nombre))
affichelignebas(tab, 10)
```

```python
moyenne = tab.mean()#permet de remplacer par la moyenne les ligne qui sont fausses
tab.fillna(moyenne)
```

```python
def convertnumerique(data, colonne, element1, element2):# cette fonction transformer la colonne Loan_Status en colonne numérique
    data.loc[data[colonne]=='element1',"colonne"]=1
    data.loc[data[colonne]=='element2',"colonne"]=2
    print(data[colonne].unique())
convertnumerique(tab,"Loan_Status", "Y","N")
```

```python

def varbinaire(data, colonne, element1, element2):# créer des colonnes représentant une variable binaire et de remplacer par la moyenne les ligne qui sont fausses
    data.loc[data[colonne]==2,"element2"]=0
    print(data[colonne].unique())
varbinaire(tab,"Loan_Status", "Y","N")
```
